Data preparation and processing

Example data:
• An experiemental drug was tested on individuals from ages 13 to 100 in a clinical trial.
⚫ The trial had 2100 participants. Half were under 65 years old, half were 65 years or older.
• Around 95% of patients 65 or older experienced side effects.
• Around 95% of patients under 65 experienced no side effects.

```python
!pip install -U scikit-learn
!pip install matplotlib
```

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler

train_labels = []
train_samples = []

for i in range(50):
    # ** The 5 % younger people who experienced side effects.
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)   # ** 1 --> patients experience side effects and 0 --> patients did not experience side effects
    
    # ** The 5 % older people who did not experience side effects.
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)

for i in range(1000):
    # ** The 95 % of younger people who did not experience side effects.
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    # ** The 95 % of older people who experienced side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
    
for j in train_samples:
    print(j)
```

```python
for k in train_labels:
    print(k)
```

As data has been generated now we can train them by using the numpy arrays of this data samples

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
train_labels, train_samples = shuffle(train_labels, train_samples)

scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaled_train_samples:
    print(i)
```

Simple tf.keras sequential model

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy

model = Sequential([
    Dense(units=16, input_shape=(1,), activation='relu'),  # ** first two lines --> hidden layer in the neural network and the last line is the output layer.
    Dense(units=32, activation='relu'),
    Dense(units=2, activation='softmax')  # ** softmax gives the probablity whether the patient experienced side effects or not.
])
```

```python
model.summary()  # ** visual representation of what we just created.
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])  # ** this is to compile the model
```

X_train and y_train are the input features and corresponding labels of the training data, respectively.
epochs specifies the number of times the entire training dataset will be passed through the network during training.
batch_size indicates the number of samples that will be used in each update of the model's parameters. The training data is divided into batches, and the model's parameters are updated after each batch.
validation_data is an optional argument that allows you to specify validation data, typically a separate set of data that is not used for training but used to evaluate the model's performance during training.

```python
model.fit(x=scaled_train_samples, y=train_labels, batch_size=10, epochs=30, shuffle=True, verbose=True)
```

validation set with Tensorflow's keras API

```python
model.fit(x=scaled_train_samples, y=train_labels, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=True)
```

```python
test_labels = []
test_samples = []

for i in range(50):
    # ** The 5 % younger people who experienced side effects.
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)   # ** 1 --> patients experience side effects and 0 --> patients did not experience side effects
    
    # ** The 5 % older people who did not experience side effects.
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(0)

for i in range(1000):
    # ** The 95 % of younger people who did not experience side effects.
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(0)
    
    # ** The 95 % of older people who experienced side effects
    random_older = randint(65,100)
    test_samples.append(random_older)
    test_labels.append(1)
    
for j in test_samples:
    print(j)
```

```python
for k in test_labels:
    print(k)
```

```python
test_labels = np.array(test_labels)
test_samples = np.array(test_samples)
test_labels, test_samples = shuffle(test_labels, test_samples)

scaler = MinMaxScaler(feature_range=(0,1))
scaled_test_samples = scaler.fit_transform(test_samples.reshape(-1,1))
```

Prediction on the test samples

```python
predictions = model.predict(x=scaled_test_samples, batch_size=10, verbose=0)

for i in predictions:
    print(i)
    
# ** output format ==> [not experiencing side effect | experiencing side effect]
```

```python
rounded_predictions = np.argmax(predictions, axis=-1)
for k in rounded_predictions:
    print(k)
```

using a confusion matrix we can visually see how well a model performed for a test data

Confusion matrix

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true=test_labels, y_pred=rounded_predictions)
```

```python

def plot_confusion_matrix(cm, classes,normalize=False,title='Confusion matrix', cmap=plt.cm.Blues):
    
    """This function prints and plots the confusion matrix. 
       Normalization can be applied by setting `normalize=True`."""

    plt.imshow(cm, interpolation='nearest', cmap=cmap) 
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    if normalize:
        cm = cm.astype('float') / cm.sum (axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')
    print(cm)
    
    thresh = cm.max()/2
    for i,j in itertools.product(range(cm.shape[0]),range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")
        
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no_side_effects','had_side_effects']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title= 'Confusion Matrix')

# ** dark blue boxes are the correct predictions
```
